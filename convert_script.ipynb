```python
import unicodedata
import pyperclip

def convert(s):
    ans = []
    for line in s.split('\n'):
        if len(line) == 0:
            ans.append(line)
        elif line in ['Aメロ', 'Bメロ', 'Cメロ', 'サビ', '大サビ', '落ちサビ', 'アウトロ', '間奏', '2番']:
            ans.append(line)
        else:
            flag = True
            for c in line:
                if unicodedata.east_asian_width(c) != 'Na':
                    ans_str = f'<span class="lylic">{line}</span>'
                    ans.append(ans_str)
                    flag = False
                    break
            if flag:
                ans.append(f'<span class="chord">{line}</span>')
    ans.insert(0, '<pre class="gakuhu">')
    ans.append('</pre>')
    print('\n'.join(ans))
    # クリップボードへコピー
    pyperclip.copy('\n'.join(ans))
```

```python
convert(s)
```
